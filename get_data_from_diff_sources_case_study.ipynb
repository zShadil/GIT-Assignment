```python
import pandas as pd
import numpy as np
```

 Step 1: Load SpaceX Launch Data from API
 1.●	Use https://api.spacexdata.com/v4/launches

```python
import requests
url = "https://api.spacexdata.com/v4/launches"
response=requests.get(url)
data=response.json()
launch_df=pd.DataFrame(data)
launch_df
```

2.	Extract relevant columns: name, date_utc, success, details, rocket

```python
launch_new = launch_df[['name','date_utc','success','details','rocket']]
launch_new
```

3	Convert date_utc to datetime and extract the year  





```python
launch_df['date_utc'] = pd.to_datetime(launch_df['date_utc'])
launch_df['year'] = launch_df['date_utc'].dt.year
launch_df.head()
```

Step 2: Load Rocket Metadata
1.	Use  https://api.spacexdata.com/v4/rockets

```python
url1='https://api.spacexdata.com/v4/rockets'
response=requests.get(url1)
data2=response.json()
rockets_df=pd.DataFrame(data2)
rockets_df
```

2.Extract id, name, type, active, and stages

```python
rockets_new = rockets_df[['id','type','active','stages']]
rockets_new
```

Step 3: Merge Launch and Rocket Data
1. Join the two DataFrames on rocket ID using   

```python
merged_df = pd.merge(launch_new,rockets_new,left_on='rocket', right_on='id', how='left')
merged_df
```

Step 4: Add Simulated Country Information
1. Add a new column country and randomly assign one of these values:
 ['USA', 'Russia', 'India', 'China', 'France']  


```python
countries = ['USA', 'Russia', 'India', 'China', 'France']
merged_df['country_simulated'] = np.random.choice(countries, size=len(merged_df))
merged_df.head()
```

Step 5: Store Merged Data in SQLite3
1.	Use sqlite3 to create a connection and save the merged DataFrame as a table named launches
2.2.Table should contain all merged columns including country

```python
import sqlite3
conn = sqlite3.connect('spacex.db')
merged_df.to_sql('launches', conn, if_exists='replace', index=False)
```

```python
df = pd.read_sql_query("SELECT * FROM launches", conn)
df.head()
```

  Step 6: Run SQL Queries on the Data to analyze
  1. Launches by Country





```python
query = """
SELECT country_simulated AS country, COUNT(*) AS launch_count
FROM launches
GROUP BY country_simulated
ORDER BY launch_count DESC;
"""

pd.read_sql_query(query, conn)
```

2.Which year had the highest number of launches?

```python
query = """
SELECT strftime('%Y', date_utc) AS year, COUNT(*) AS launch_count
FROM launches
GROUP BY year
ORDER BY launch_count DESC
LIMIT 1;
"""

pd.read_sql_query(query, conn)
```

3. Top 5 Missions by Launch Count

```python
query = """
SELECT name AS mission_name, COUNT(*) AS launch_count
FROM launches
GROUP BY mission_name
ORDER BY launch_count DESC
LIMIT 5;
"""

pd.read_sql_query(query, conn)
```
