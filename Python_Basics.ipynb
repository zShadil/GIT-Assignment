```python
# Variables and data types to store account information [cite: 4]
account_holder = "John Doe"
balance = 1000.0

def balance_enquiry():
    """Function to perform balance inquiry [cite: 6]"""
    print("--- Current Balance ---")
    print(f"Account Holder: {account_holder}")
    print(f"Balance: ${balance}") # Display the output [cite: 15]

def deposit(amount):
    """Function to perform deposit [cite: 6]"""
    global balance
    balance = balance + amount
    print(f"Action: Deposited ${amount}") # Display the output [cite: 15]
    print(f"New Balance: ${balance}")

def withdraw(amount):
    """Function to perform withdrawal [cite: 6]"""
    global balance
    print(f"Action: Attempting to withdraw ${amount}")

    # Conditional statements to check for sufficient balance [cite: 5]
    if amount > balance:
        print("Error: Insufficient balance for withdrawal.") # Display the output [cite: 15]
    else:
        balance = balance - amount
        print(f"Withdrawal successful. Remaining Balance: ${balance}") # Display the output [cite: 15]

# Running the operations to show output
balance_enquiry()
deposit(500)
withdraw(200)
withdraw(2000) # This will trigger the "insufficient balance" check
```

```python
# Variables and data types to store weather data [cite: 10]
temperatures = [32, 28, 30, 25, 29]
humidities = [45, 50, 40, 60, 55]

def analyze_weather(temps, hums):
    """Function to calculate averages and print results """

    # Calculate average temperature and humidity
    avg_temp = sum(temps) / len(temps)
    avg_hum = sum(hums) / len(hums)

    # Conditional statements to determine weather conditions [cite: 11]
    if avg_temp > 28:
        condition = "Sunny"
    elif avg_hum > 70:
        condition = "Rainy"
    else:
        condition = "Cloudy"

    # Display the results [cite: 13, 15]
    print("--- Weather Analysis Report ---")
    print(f"Average Temperature: {avg_temp}")
    print(f"Average Humidity: {avg_hum}")
    print(f"Predicted Condition: {condition}")

# Run the function
analyze_weather(temperatures, humidities)
```
