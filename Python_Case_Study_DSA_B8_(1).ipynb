# Q1. Write a function that return the range (maximum and minimun) of a given list of numbers

```python
def find_range(numbers):
    minimum_value = numbers[0]
    # setting the first value of the list as minimum value
    maximum_value = numbers[0]
    # setting the first value of the list as maximum value

    for number in numbers:
        # looping through the entire list to check for
        # maximum_value and minimum_value
        if (number >= maximum_value):
            maximum_value = number
        if (number <= minimum_value):
            minimum_value = number

    return(minimum_value, maximum_value)

list =[12,34,67,89,90]
range = find_range(list)
range



```

## Q2. Write function that returns mean value of a list of numbers

```python
def find_mean(numbers):
    total = 0
    # variable to store the sum of all numbers

    count = 0
    # variable to count how many numbers are in the list

    for number in numbers:
        total += number
        count += 1
    mean_value = total / count
    return mean_value
print(find_mean([5,10,15]))
```
