```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
iris
```

```python
#Line plot
plt.plot( iris['sepal_length'])

plt.xlabel("x values")
plt.ylabel("Sepal Length")
plt.title("Line Plot of Sepal Length")
plt.show()
```

```python
plt.plot(iris['sepal_length'],color='green')
plt.plot(iris['sepal_width'],color='blue')
plt.plot(iris['petal_length'],color='purple')
plt.plot(iris['petal_width'],color='teal')

plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Line Plot of Iris Features")
plt.show()
```

```python
plt.plot(iris['sepal_length'],color="purple",marker='o',linestyle='dashed',markerfacecolor='pink',markersize=10)
```

```python
plt.plot(iris['sepal_width'],color='olive')

avg_sepal_length = np.mean(iris['sepal_width'])

plt.axhline(y=avg_sepal_length, color='black', linestyle='--')

plt.xlabel('x values')
plt.ylabel('Sepal width')
plt.title('Sepal width with Average Line')
plt.legend(['Sepal width', 'Average Sepal width'])
plt.show()
```

```python
#Bar plot
plt.bar(iris["species"], iris["sepal_length"],color="purple")
plt.xlabel("Species")
plt.ylabel(" Sepal Length")
plt.title(" Sepal Length of Iris Species")
plt.show()
```

```python
sepal_width = iris.groupby("species")["sepal_width"].first()
petal_width  = iris.groupby("species")["petal_width"].first()
plt.bar(sepal_width.index,sepal_width.values, color="orange")
plt.bar(petal_width.index, petal_width.values, color="r")
```

```python
#scatter plot
sns.scatterplot(data=iris,x="sepal_length",y="sepal_width",hue="species")
```

```python
#piechart
iris_species=iris['species'].value_counts()
labels=iris_species.index
plt.pie(iris_species,labels=labels)
plt.show()
```
